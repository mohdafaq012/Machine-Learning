Importing the Dependecies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection & Preprocessing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
X = data_frame.drop(columns = 'label', axis = 1)      # column wise drop
Y = data_frame['label']                               # axis = 0 for row drop
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)        # 20% test 80%train, random state is to do same as others
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)       # fit and transform with train data
X_test_std = scaler.transform(X_test)             # only transform to the test data
```

```python
print(X_train_std)
```

# Building the Neural Networkimage.png

Importing TensorFlow and Keras

```python
import tensorflow as tf
tf.random.set_seed(3)                 # this will give us same as others    # same things and weights and biases again and again
from tensorflow import keras
```

Setting up the layers of the Neural Network

```python
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),           # one input layer   ( flatten is used for converting the data into single dimensional array )
                          keras.layers.Dense(20, activation='relu'),         # one hidden layer
                          keras.layers.Dense(2, activation='sigmoid')       # one output layer    (2 neuron for each output)
])
```

```python
# Compiling the Neural Network
model.compile(optimizer = 'adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# training the Neural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=25)    # validation split is dev here....
```

Visualising accuracy and loss


```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training_data', 'validation_data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model_loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training_data', 'validation_data'], loc='upper right')
```

Accuracy of the mdoel on the test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
input_data = (12.18,20.52,77.22,458.7,0.08013,0.04038,0.02383,0.0177,0.1739,0.05677,0.1924,1.571,1.183,14.68,0.00508,0.006098,0.01069,0.006797,0.01447,0.001532,13.34,32.84,84.58,547.8,0.1123,0.08862,0.1145,0.07431,0.2694,0.06878)
# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# Standardising the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if prediction_label[0] == 0:
  print("The breast cancer is Malignant")
else:
  print("the breast cancer is Beging")
```
